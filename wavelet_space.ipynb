# Wavelet Module testing/debugging

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import data
import pytorch_wavelet as wavelet

%matplotlib notebook
%load_ext autoreload
%autoreload 2

dtype = torch.complex64
ctype = torch.float32
device = 'cpu'
```

## Load Image

```python
x = data.camera()
x = torch.from_numpy(x)
```

## Wavelet Transform testing

```python
a1 = wavelet.visualize(x, Nlayers=1)
a2 = wavelet.visualize(x, Nlayers=2)

plt.figure()
plt.subplot(131)
plt.imshow(x)
plt.axis('off')
plt.title('Image')
plt.subplot(132)
plt.imshow(a1)
plt.axis('off')
plt.title('Wavelet Level 1')
plt.subplot(133)
plt.imshow(a2)
plt.axis('off')
plt.title('Wavelet Level 2')

plt.savefig('figures/output.pdf', dpi=300, transparent=True)
```
